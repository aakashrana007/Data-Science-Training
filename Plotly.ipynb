### Create a function `compare_countries` that produces similar chart comparing two countries.
Eg: `compare_countries('India','Nepal')`

```python
def compare_countries(country1,country2):
  df_country=data.query(f"country in ['{country1}','{country2}']")
  fig=px.line(df_country, x='lifeExp', y='gdpPercap',color='country',text='pop')
  fig.update_traces(textposition='bottom right')
  fig.show()

compare_countries('China','India')
```

#Geojson

GeoJSON is a format for encoding a variety of geographic data structures using JavaScript Object Notation (JSON). Plotly, a graphing library for Python and other programming languages, supports GeoJSON for creating geographic visualizations, such as maps with markers, polygons, and other geographical features.

```python
import plotly.express as px

df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth_mapbox(df,
                           geojson=geojson,
                           color="Joly",
                           locations="district",
                           featureidkey="properties.district",
                           center={"lat": 45.5517, "lon": -73.7073},
                           mapbox_style="carto-positron",
                           zoom=9)
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
fig = px.scatter_geo(df,
                     locations="iso_alpha",
                     color="continent",
                     hover_name="country",
                     size="pop",
                     animation_frame="year",
                     projection="natural earth"
                    )
fig.show()
```
