```python

age_df=df.groupby(by="Country")["Age"].mean().sort_values(ascending=False).reset_index(name='Avg_Age')
age_df
```

```python
#AVERAGE AGE DISTRIBUTION OF COUNTRIES
plt.figure(figsize=(25,8))
plt.title("Average Age of each Country", fontdict = {'fontname':'DejaVu Sans', 'fontsize':20, 'color': 'black'})
plt.xlabel('Country', fontdict = {'fontsize':30})
plt.ylabel('age mean')
sns.barplot(x='Country', y='Avg_Age', data=age_df, palette='viridis')
```

```python
caps_df=df.groupby(by="Country")["Caps"].mean()
caps_df=caps_df.sort_values(ascending=False)
```

```python
import plotly.express as px

quarter_finalists = ['France', 'England', 'Germany', 'Spain', 'Portugal', 'Switzerland', 'Turkiye', 'Netherlands']
attackers_positions = ['Centre-Forward', 'Left Winger', 'Right Winger', 'Second Striker', 'Central Midfield', 'Defensive Midfield', 'Attacking Midfield']
attackers = df[(df['Country'].isin(quarter_finalists)) & (df['Position'].isin(attackers_positions))]


fig = px.scatter(attackers,
              x='Age',
              y='Caps',
              color='Name',
              animation_frame='Country',  # Use the country for animation
              animation_group=attackers['Name'],
              size='MarketValue',
              text='Name',  # Add player names as text
              title='Caps vs Age for Quarter-Finalist Countries',
              size_max=35,
              range_x=[15,45],
              range_y=[0,200]
             )

# Add dots for each player
fig.update_traces(textposition='top center')

# Show the plot
fig.show()
```
