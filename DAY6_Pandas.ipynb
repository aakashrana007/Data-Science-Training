# **Pandas-**
-used to analyze data

```python
import pandas as pd
```

```python
#creating a data set
people={
    'first':['Ram','Shyam','Gopal'],
    'last':['Sharma','Yadav','Verma'],
    'email':['ram@gmail.com','hari@yahoo.com','gopal@hotmail.com']
    }
  #to access email
people['email']
people.get('email')
```

Data Set-

```python
people_df=pandas.DataFrame(people)
people_df
```

```python
people_df['email']   #access the email
```

```python
dir(people_df)      #check what can you do with pandas
```

```python
#to access multiple column
people_df[['email','last']]
```

```python
#returns first n rows; by default it prints 5 but we can give index ourselves
people_df.head(1)
```

```python
#returns the last n rows; by default 5
people_df.tail(2)
```

```python
#gives information of all data
people_df.info()
```

```python
#gives info for interger data like sum min max frequency
people_df.describe()
```

```python
#lists all the columns
people_df.columns
```

To access elements-
1. loc- based on label/name

2. iloc- based on index

```python
people_df.iloc[(0,2)]         #access row 0's column 2's elem
```

```python
people_df.iloc[[0,2],[0,2]]     #[row,column]
```

```python
people_df.loc[[0,2],['first','email']]    #access first and email column of index 0 and 2
```

```python
people_df.loc[0:1]        #slicing; prints row of index 0 to 1
```

Access 3rd to 7th row with iloc

Access 4th to 8th row with loc

Access the first and last name of 4th index
using loc and iloc both

Access first, last, and email of 5 to 7th index
using loc and iloc both

```python
dataset = {
    'first': [
        'Ram', 'Hari', 'Gopal', 'Isabella', 'Oliver', 'Sophia', 'Elijah', 'Emma',
        'William', 'Ava'
    ],
    'last': [
        'Sharma', 'Yadav', 'Verma', 'Smith', 'Johnson', 'Brown', 'Davis', 'Miller',
        'Wilson', 'Moore'
    ],
    'email': [
        'ram@gmail.com', 'hari@yahoo.com', 'gopal@hotmail.com', 'isabella@example.com',
        'oliver@example.com', 'sophia@example.com', 'elijah@example.com', 'emma@example.com',
        'william@example.com', 'ava@example.com'
    ]
}

dataset_df = pd.DataFrame(dataset)
dataset_df

```

```python

#Access 3rd to 7th row with iloc
dataset_df.iloc[2:8]
```

```python
#Access 4th to 8th row with loc
dataset_df.loc[3:8]
```

```python
#Access the first and last name of 4th index using loc
dataset_df.loc[3,['first','last']]
```

```python
#using iloc
dataset_df.iloc[3,0:2]
```

```python
#Access first, last, and email of 5 to 7th index using loc and iloc both
dataset_df.loc[5:7,['first','last','email']]
```

```python
#using iloc
dataset_df.iloc[5:8]
```
