```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,6]
plt.title("our first graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6 ]
plt.title("our second graph", {'fontsize':25, 'color':'red'})
plt.xlabel("x-axis", {'fontsize':25, 'color':'blue'})
plt.ylabel("y-axis",{'fontsize':25, 'color':'orange'})
plt.plot(x,y, color='green')
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6 ]
plt.title("our second graph", fontdict= {'fontsize':25, 'color':'red'})
plt.xlabel("x-axis", fontdict= {'fontsize':25, 'color':'blue'})
plt.ylabel("y-axis",fontdict = {'fontsize':25, 'color':'orange'})
plt.plot(x,y, color='green')
plt.show()
```

- Can you add plot title `"Petrol prices over time in different countries"`?
- In the `fontdict`, can you set the `fontname` to `sans-serif`?
- Can you change the `fontsize` to `20` and `color` to `#d08d88`?
- Can you change the `xlabel` to `"Years"`?
- Can you change the `ylabel` to `"Price in dollars"`?
- Plot the values of `USA`, `Japan`, and `Australia` with `markers` and `labels`.
- Show the legend as well.

```python
list=['USA','Japan','Australia']
sign=['*','.','^']
for i in list:
   plt.plot(petrol['Year'],petrol[i],label=i,marker=sign[list.index(i)])
plt.title("Petrol prices over time in different countries",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.xlabel("Years",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.ylabel("Price in dollars",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.legend()
plt.show()
```
