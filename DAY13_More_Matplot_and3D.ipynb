```python
petrol=pd.read_csv('/content/drive/MyDrive/data/petrol_prices.csv')
petrol
```

```python
list=['USA','Japan','Australia']
sign=['*','.','^']
for i in list:
   plt.plot(petrol['Year'],petrol[i],label=i,marker=sign[list.index(i)])
plt.title("Petrol prices over time in different countries",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.xlabel("Years",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.ylabel("Price in dollars",fontdict={'fontname':'sans-serif','fontsize':20,'color':'#d08d88'})
plt.legend()
plt.show()
```

```python
x=petrol['Year']
plt.figure(figsize=(15,8))  #set the figure ourselves; 15x8
for i in petrol.columns[1:]:
  plt.plot(petrol['Year'],petrol[i],label=i,marker='x')

plt.xticks(x)
plt.xlabel('Year')
plt.ylabel('Price in dollars')
plt.title('Petrol prices over time in different countries')
plt.legend()
plt.show()
```

**Bar Graph-**

```python
x=['HR','QA','Developer']
y=[40000,50000,90000]
bars=plt.bar(x,y)

bars[0].set_hatch('x')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
bars[2].set_color('r')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.title('Salary of employees')
```

**Exercise-**

* Change the index to 'Year'
* Create a function 'year_plot'
(calling year_plot(2001) should give the plot of the year 2001.)

```python
petrol_df=pd.read_csv('/content/drive/MyDrive/data/petrol_prices.csv', index_col='Year')
petrol_df
```

```python
def year_plot(year):
  if year in petrol_df.index:
    plt.title(f"The plot for {year}")
    plt.bar(petrol_df.columns,petrol_df.loc[year])
    plt.figure(figsize=(20,10))
  else:
    print("Data not found")

year_plot(2001)
```
